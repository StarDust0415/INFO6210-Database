```python
import pandas as pd
from pymongo import MongoClient
import json

def mongoimport(csv_path, db_name, coll_name, db_url='localhost', db_port=27017):
    client = MongoClient(db_url, db_port)
    db = client[db_name]
    coll = db[coll_name]
    data = pd.read_csv(csv_path, encoding='latin1')
    payload = json.loads(data.to_json(orient='records'))
    coll.remove()
    coll.insert(payload)
    return coll.count()
```

### Q1:What are tags are associated with a person, place or thing?

```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
```

```python
db = client.local
collection=db.soccer_twitter
```

```python
pprint.pprint(collection.find_one({"player_name": "Lionel Messi"}))
```

```python
list(db.soccer_twitter.aggregate([
    {"$match":{"player_name":"Lionel Messi"}},
    {"$group":{"_id":"$array_agg","count":{"$sum":1}}},
    {"$sort":{"count":-1}}]))
```

### Q2:What social media users are like other social media users in your domain?

```python
match = {
    'tweets':{"$regex": "Lionel Messi"}
}

group = {
    '_id':"$user_id"
}

ret2 = db.soccer_twitter.aggregate([
    {'$match':match},
    {'$group': group}
])

list(ret2)
```

```python
match = {
    'tweets':{"$regex": "Cristiano Ronaldo"}
}

group = {
    '_id':"$user_id"
}

ret2 = db.soccer_twitter.aggregate([
    {'$match':match},
    {'$group': group}
])

list(ret2)
```

### Q3:What people, places or things are popular in your domain?

```python
list(db.soccer_twitter.aggregate([
    {"$group":{"_id":"$array_agg","count":{"$sum":1}}},
    {"$sort":{"count":-1}}]))
```

```python
for doc in collection.find({},{"player_name":1, "_id":0, "retweet_count":1}).sort('retweet_count', pymongo.DESCENDING):
    print(doc)
```

### Q4:What people, places or things are trending in your domain? (A trend is popularity over time.)

```python
list(db.soccer_twitter.aggregate([
    {"$sort":{"created_at":1}}]))
```
