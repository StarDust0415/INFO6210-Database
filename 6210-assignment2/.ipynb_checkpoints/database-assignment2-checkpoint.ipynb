## search soccer players and get tweets and hashtags about them

```python
#Getting Tweets
import json

search_words = ['#Lionel Messi', '#Cristiano Ronaldo', '#Neymar', '#Wayne Rooney', '#Mario Balotelli', '#Andrea Pirlo', '#Gareth Bale','#Robert Lewandowski']
search_team = ['Lionel Messi', 'Cristiano Ronaldo', 'Neymar', 'Wayne Rooney', 'Mario Balotelli', 'Andrea Pirlo', 'Gareth Bale','Robert Lewandowski']
q = 'EuropeanSoccer' 
n = 105
from urllib.parse import unquote
search_results = api.search(q = q, count = n)
```

```python
USER = "postgres"
PASSWORD = ""
DATABASE = "postgres"
```

```python
#define functions of getting tweets and write it to the postgresql database
def create_twitter_table():
    conn = psycopg2.connect(database=DATABASE,user=USER, password=PASSWORD)
    cur = conn.cursor()
    cur.execute('''CREATE TABLE EuropeanSoccer_twitter(screen_name varchar,created_at varchar,text varchar)''')
    conn.commit()
    conn.close()
```

```python
def write_tweets_sql(created_at, text, screen_name):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWORD)
    cursor = db.cursor()
    insert_query = "INSERT INTO EuropeanSoccer_twitter(screen_name,created_at,text) VALUES ( %s, %s, %s)"
    cursor.execute(insert_query, (screen_name, created_at, text))
    db.commit()
    cursor.close()
    db.close()
    return
def store_tweets():
    for word in search_words:
        for tweet in api.search(q = word, count = 100):
            write_tweets_sql(tweet.created_at,tweet.text,tweet.user.screen_name)
            print(tweet.created_at,tweet.text, tweet.user.screen_name)
```

```python
create_twitter_table()
```

```python
store_tweets()
```

```python
##define functions of getting hashtags and write it to the postgresql database
def create_hashtag_table():
    conn = psycopg2.connect(database=DATABASE,user=USER, password=PASSWORD)
    cur = conn.cursor()
    cur.execute('''CREATE TABLE EuropeanSoccer_tags(tag varchar, team varchar)''')
    conn.commit()
    conn.close()
```

```python
def write_tags_sql(tag,movie):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWORD)
    cursor = db.cursor()
    insert_query = "INSERT INTO EuropeanSoccer_tags(tag, team) VALUES ( %s, %s)"
    cursor.execute(insert_query, (tag,movie))
    db.commit()
    cursor.close()
    db.close()
    return
```

```python
def store_hashtag():
    for i in range (0,9):
        word=search_words[i];
        team=search_team[i];
        for tweet in api.search(q = word, count = 100):
            for hashtag in tweet.entities.get('hashtags'):
                write_tags_sql(hashtag.get('text'),team)
                print(hashtag, team)
```

```python
user1 = api.get_user('footy90com')
print(user1.name, user1.screen_name, str(user1.statuses_count), str(user1.friends_count), str(user1.followers_count))
```

```python
from datetime import datetime, date, time, timedelta
tweets = user1.statuses_count
account_created_date = user1.created_at
delta = datetime.utcnow() - account_created_date
account_age_days = delta.days
print("Account age (in days): " + str(account_age_days))
if account_age_days > 0:
    print("Average tweets per day: " + "%.2f"%(float(tweets)/float(account_age_days)))
```

```python
from tweepy import Cursor
hashtags = []
mentions = []
tweet_count = 0
end_date = datetime.utcnow() - timedelta(days=30)
for status in Cursor(api.user_timeline, id='footy90com').items():
      tweet_count += 1
      if hasattr(status, "entities"):
        entities = status.entities
        if "hashtags" in entities:
          for ent in entities["hashtags"]:
            if ent is not None:
              if "text" in ent:
                hashtag = ent["text"]
                if hashtag is not None:
                  hashtags.append(hashtag)
        if "user_mentions" in entities:
          for ent in entities["user_mentions"]:
            if ent is not None:
              if "screen_name" in ent:
                name = ent["screen_name"]
                if name is not None:
                  mentions.append(name)
      if status.created_at < end_date:
        break
```

# Trending

```python
import time
class MyStreamListener(tweepy.StreamListener):
    def __init__(self, time_limit=60):
        self.start_time = time.time()
        self.limit = time_limit
        self.saveFile = open('messi.json', 'a')
        super(MyStreamListener, self).__init__()

    def on_data(self, data):
        if (time.time() - self.start_time) < self.limit:
            self.saveFile.write(data)
            #self.saveFile.write('\n')
            return True
        else:
            self.saveFile.close()
            return False

myStream = tweepy.Stream(auth=api.auth, listener=MyStreamListener(time_limit=2000))
myStream.filter(track=['#messi'])
```

```python
import sys 
from datetime import datetime 
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import numpy as np
import pickle
```

```python
import json

file1 = '/Users/yizheliu/Desktop/6210-assignment2/messi.json'
with open(file1) as f:
    all_dates = []
    for line in f:
        tweet = json.loads(line)
        all_dates.append(tweet.get('created_at'))
    idx = pd.DatetimeIndex(all_dates)
    ones = np.ones(len(all_dates))
    # the actual series (at series of 1s for the moment) 
    my_series = pd.Series(ones, index=idx)
    
    # Resampling / bucketing into 1-minute buckets
    #per_minute = my_series.resample('1Min', how='sum').fillna(0)
    per_minute = my_series.resample('1Min').sum()
    # Plotting the series
    fig, ax = plt.subplots() 
    ax.grid(True) 
    ax.set_title("Tweet Frequencies")
    hours = mdates.MinuteLocator(interval=20) 
    date_formatter = mdates.DateFormatter('%H:%M')
    
    datemin = datetime(2018, 3, 18, 20, 20) 
    datemax = datetime(2018, 3, 18, 21, 0)
    
    ax.xaxis.set_major_locator(hours) 
    ax.xaxis.set_major_formatter(date_formatter) 
    ax.set_xlim(datemin, datemax)
    max_freq = per_minute.max() 
    ax.set_ylim(0, max_freq) 
    ax.plot(per_minute.index, per_minute)
    
    plt.savefig('tweet_time_series.png')
```
